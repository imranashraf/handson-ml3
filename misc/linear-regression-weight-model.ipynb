# Scikit Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("height-weight.csv")
df.head()
```

# Height to Weight Linear Model


```python
X = df.Height.values
y = df.Weight.values

plt.scatter(X, y, alpha=0.3)
plt.title('Weight - Height Relationship')
plt.xlabel('Height in meters')
plt.ylabel('Weight in kgs')
plt.show()
```

```python
# Un-comment the following to install scikit-learn
# !pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression

X = X.reshape(-1, 1)

regression = LinearRegression()
regression.fit(X, y)
```

```python
h = 1.6
w_Pred = regression.predict([[h]])
print('weight of ', h, 'm tall person is predicted to be: ', w_Pred)
```

```python
plt.scatter(X, y, alpha=0.3)
plt.plot(X, regression.predict(X), color='red', linewidth=3)

# plot the predicted value
plt.plot(h, w_Pred, marker="o")
plt.title('Height - Weight Relationship')
plt.xlabel('Height in m')
plt.ylabel('Weight in kgs')
plt.show()
```

# Obtain slope and intercept

```python
print(regression.intercept_)
print(regression.coef_)
```

```python
b = regression.intercept_
m = regression.coef_[0]

def calculate_weight(ht):
    return b + ht * m

print('weight of ', h, 'm tall person is predicted to be: ', calculate_weight(h))
```

### Pearson correlation coefficient
The Pearson correlation coefficient, often denoted by $r$, a (common) measure of linear dependence between two variables. 





- When $r=1$, we have a perfect positive linear correlation, that is, an increase of one variable predicts an increase of the other. 
- When $r=-1$, we have a perfect negative linear correlation and the increase of one variable predicts a decrease of the other. 
- When $r=0$, we have no linear correlation.



As a general rule, we will get intermediate values.

**Note:** Pearson correlation coefficient says nothing about non-linear correlations.
